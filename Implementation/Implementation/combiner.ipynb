```python
import scipy.io
from scipy.io import loadmat
from scipy.io import savemat
import numpy as np
import pandas as pd
```

```python
def combine_mat_files(file1_path, file2_path,file3_path, output_file):

    # Load data from the first file
    mat_data1 = scipy.io.loadmat(file1_path)
    array1 = mat_data1['all_features']  

    # Load data from the second file
    mat_data2 = scipy.io.loadmat(file2_path)
    array2 = mat_data2['all_features']  

    mat_data3 = scipy.io.loadmat(file3_path)
    array3 = mat_data3['all_features']  

    # Combine the two arrays into a single row
    combined_array = np.concatenate((array1, array2,array3), axis=1)
    print(combined_array.shape)

    # Save the combined array into a new .mat file
    combined_data = {'all_features': combined_array} 
    scipy.io.savemat(output_file, combined_data)

```

```python
# Example usage
file1_path = 'cover_1_4.mat'  
file2_path = 'cover_4_8.mat'  
file3_path = 'cover_8_10.mat'
output_file = 'cover_10k.mat'  # Specify the output file name

```

```python
combine_mat_files(file1_path, file2_path,file3_path,output_file)
```
